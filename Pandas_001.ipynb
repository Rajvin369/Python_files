```python
# Series Methods
#----------------

# .head(): Returns the first n rows.

# .tail(): Returns the last n rows.

# .describe(): Generates descriptive statistics.

# .mean(): Returns the mean.

# .median(): Returns the median.

# .std(): Returns the standard deviation.

# .sum(): Returns the sum.

# .cumsum(): Returns the cumulative sum.




# DataFrame Creation
#-------------------

# pd.DataFrame(): Creates a DataFrame.

# pd.read_csv(): Reads a CSV file into a DataFrame.

# pd.read_excel(): Reads an Excel file into a DataFrame.

# pd.read_sql(): Reads a SQL query or database table into a DataFrame.

# pd.read_json(): Reads a JSON file into a DataFrame.




# DataFrame Manipulation
#------------------------


# .drop(): Removes rows or columns.

# .sort_values(): Sorts by values along either axis.

# .sort_index(): Sorts by index.

# .groupby(): Groups DataFrame using a mapper or by series of columns.

# .merge(): Merges DataFrames.

# .concat(): Concatenates DataFrames.

# .pivot_table(): Creates a pivot table.




# Data Selection
#----------------

# .loc[]: Access a group of rows and columns by labels.

# .iloc[]: Access a group of rows and columns by integer positions.

# .at[]: Access a single value for a row/column label pair.

# .iat[]: Access a single value for a row/column pair by integer position.




# Data Cleaning
#----------------

# .dropna(): Removes missing values.

# .fillna(): Fills missing values.

# .replace(): Replaces values.

# .duplicated(): Finds duplicate rows.

# .drop_duplicates(): Removes duplicate rows.



# Statistical Functions
#-----------------------

# .mean(): Returns the mean of the values.

# .median(): Returns the median of the values.

# .std(): Returns the standard deviation of the values.

# .var(): Returns the variance of the values.

# .corr(): Returns correlation of a DataFrame.




# Time Series
#------------

# .to_datetime(): Converts argument to datetime.

# .resample(): Resamples time-series data.

# .shift(): Shifts index by desired number of periods.

# .rolling(): Provides rolling window calculations.
```

```python
# Example DataFrames


left = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K3'],
                     'A': ['A0', 'A1', 'A2', 'A3'],
                     'B': ['B0', 'B1', 'B2', 'B3']})

right = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K3'],
                          'C': ['C0', 'C1', 'C2', 'C3'],
                          'D': ['D0', 'D1', 'D2', 'D3']})
```

```python

left

```

```python
right
```

```python
# Merging

pd.merge(left,right,how='inner',on='key')
```

```python
left = pd.DataFrame({'key1': ['K0', 'K0', 'K1', 'K2'],
                     'key2': ['K0', 'K1', 'K0', 'K1'],
                        'A': ['A0', 'A1', 'A2', 'A3'],
                        'B': ['B0', 'B1', 'B2', 'B3']})

right = pd.DataFrame({'key1': ['K0', 'K1', 'K1', 'K2'],
                               'key2': ['K0', 'K0', 'K0', 'K0'],
                                  'C': ['C0', 'C1', 'C2', 'C3'],
                                  'D': ['D0', 'D1', 'D2', 'D3']})
```

```python
pd.merge(left, right, on=['key1', 'key2'])
```

```python
pd.merge(left, right, how='outer', on=['key1', 'key2'])
```

```python
pd.merge(left, right, how='right', on=['key1', 'key2'])
```

```python
pd.merge(left, right, how='left', on=['key1', 'key2'])
```

```python
# Joining


left = pd.DataFrame({'A': ['A0', 'A1', 'A2'],
                     'B': ['B0', 'B1', 'B2']},
                      index=['K0', 'K1', 'K2'])

right = pd.DataFrame({'C': ['C0', 'C2', 'C3'],
                    'D': ['D0', 'D2', 'D3']},
                      index=['K0', 'K2', 'K3'])
```

```python
left.join(right)
```

```python
left.join(right, how='outer')
```
